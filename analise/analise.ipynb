```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test1.csv')
```

```python
len(df)
```

```python
df = df.drop(['DEBUG:root:'],axis=1)
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,8))
plt.title('Tempo de Espera Grafo 1')
df.WAIT.plot()
plt.savefig('img.png')
```

```python
df['Time_d'] = df.WAIT/df.QTE
```

```python
plt.figure(figsize=(20,8))
plt.title('Tempo Medio de Espera Grafo 1')
df.Time_d.plot()
plt.savefig('tempopordrink.png')
```
